```python
import matplotlib.pyplot as plt

# Assuming some utility values for each food item
utilities = {'Taco': 100, 'Burger': 50, 'Salad': 30}

# Defining the lotteries
lotteries = {
    'A': [('Taco', 1)],
    'B': [('Burger', 0.25), ('Salad', 0.75)],
    'C': [('Burger', 0.2), ('Taco', 0.8)],
    'D': [('Burger', 0.4), ('Salad', 0.6)]
}

# Creating a plot
fig, ax = plt.subplots()

# Plotting each lottery
for name, lottery in lotteries.items():
    for outcome, prob in lottery:
        ax.plot(prob, utilities[outcome], 'o', label=f'{name}: ({outcome}, {prob})')

# Adding labels and legend
ax.set_xlabel('Probability')
ax.set_ylabel('Utility')
ax.legend()

# Showing the plot
plt.show()
```
